```python
# Path to where the CSV files are
dataset_path = r"C:\Users\Admin\.cache\kagglehub\datasets\ellipticco\elliptic-data-set\versions\1\elliptic_bitcoin_dataset"
```

```python
# Load features and labels
features = pd.read_csv(os.path.join(dataset_path, "elliptic_txs_features.csv"), header=None)
labels = pd.read_csv(os.path.join(dataset_path, "elliptic_txs_classes.csv"))


print(features.columns)
print(labels.columns)
```

```python
feature_columns = ['txId'] + [f'feature_{i}' for i in range(1, features.shape[1])]
features.columns = feature_columns
```

```python
print(features.columns)
print(labels.columns)
```

```python
# Merge with labels
data = pd.merge(features, labels, on='txId')
```
